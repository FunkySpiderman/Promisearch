### Definición del muestreo

El número de bins es 19.

El número de páginas que queremos por bins es 10.

El número de repos que queremos por página es 10.

Nueva división:

* 10 a 20: uno por uno.
* 20 a 30: dos por dos.
* 30 a 40: un bin.
* 40 a 70: un bin.
* 70 a 160: un bin.
* 160 a $\infty$: un bin.

```python
stars_list = []
resultados = []
repos_final = {}
rangos_estrellas = list(zip(min_estrellas_lista, max_estrellas_lista))

for min_estrellas, max_estrellas in rangos_estrellas:
    repos_bins = []
    rand_pags=random.sample(list(range(100)),10)
    i=0
    while i <10:
        # página aleatoria
        num_pagina = rand_pags[i]

        # Construyo la URL
        url_objetivo = get_pages_repos(min_estrellas, max_estrellas, num_pagina)
        
        # Aqui hago la peticion a github
        # search_github = requests.get(url_objetivo)
        search_github = requests.get(url_objetivo ,headers=headers)
        #print(search_github)
        search_github_json = search_github.json()

        try:
            lista_repos = search_github_json['items']
            lista_random = random.sample(lista_repos, 10)
            repos_bins.extend(lista_random)
            print(f'Todo bien {_+1}')
            resultados.append(search_github_json['total_count'])
            stars_list.append([min_estrellas, max_estrellas])
        except:  
            print(search_github_json)
            time.sleep(61)
            i-=1
        i+=1

    # print(repos_bins)
    repos_final[f'{min_estrellas}-{max_estrellas}'] = repos_bins
    collecion_repos.insert_one({f'{min_estrellas}-{max_estrellas}':repos_bins})
```
